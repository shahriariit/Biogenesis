```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=[]
tablename= ["maximum_cf.csv","avg_cf.csv","maximum_dt.csv", "avg_dt.csv"]

for i,j in zip(range(len(tablename)),tablename):
     li = pd.read_csv(j)
     data.append(li)
```

```python
data[1]
```

```python
line_color=['green','red','blue','yellow','orange','magenta','cyan','black']
```

```python
def xValueLoad(x):
   vals = [] 
   for i in range(len(x.columns)):
      val = x.iloc[0:10,i] 
      vals.append(val)
    
   vals.remove(vals[0])
   return vals;
```

```python
x_list=[]

for i in range(4):
  temp=xValueLoad(data[i])
  x_list.append(temp)
```

```python
x_list[0]
```

```python
y_list=[['DT','GB','KNN','LR','MLP','NB','RF','SVM','XGB','Average'], 
        ['DT','GB','KNN','LR','MLP','NB','RF','SVM','XGB','Average'], 
        ['DS01','DS02','DS03','DS04','DS05','DS06','DS07','DS08','DS09','Average'],
        ['DS01','DS02','DS03','DS04','DS05','DS06','DS07','DS08','DS09','Average']]
```

```python
SMALL_SIZE = 22
MEDIUM_SIZE = 20
BIGGER_SIZE = 22

plt.rc('font', size=SMALL_SIZE)          # controls default text sizes
plt.rc('axes', titlesize=SMALL_SIZE)     # fontsize of the axes title
plt.rc('axes', labelsize=SMALL_SIZE)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('ytick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('legend', fontsize=SMALL_SIZE)    # legend fontsize
plt.rc('figure', titlesize=BIGGER_SIZE)  # fontsize of the figure title
```

```python
def graphFunc(l1,y,l2,lt,col,bboxanch):
  plt.figure(figsize=(28,15))
    
  for i,j,k in zip(l1,l2,l2):
     plt.plot(y,i, color=j, linestyle='solid', linewidth = 3, marker='o', markerfacecolor=k, markersize=12)
  
  plt.xlabel('Classifiers') 
  plt.ylabel('Results') 
  plt.legend(('Accuracy', 'AUC', 'F-Measure','G-Mean','Sensitivity','Specificity','Type I Error','Type II Error'),loc=lt,ncol=col, bbox_to_anchor=bboxanch)   
  plt.show()
```

```python
graphFunc(x_list[0],y_list[0],line_color,'center left',2, (0.1,0.47,0,0))
```

```python
graphFunc(x_list[1],y_list[1],line_color,'center left',2,(0,0.5,0,0))
```

```python
graphFunc(x_list[2],y_list[2],line_color,'center',1,(0.5, 0.5))
```

```python
graphFunc(x_list[3],y_list[3],line_color,'best',1,(0.2, 0.5))
```
